```python
from scipy import stats
```

```python
B = stats.binom(20, 0.3)
```

```python
B.std()
```

```python
rvs = B.rvs (size = 100000)
rvs
```

```python
import matplotlib.pyplot as plt
plt.hist(rvs)
plt.show()
```

```python
P = stats.poisson (mu=15)
rvs = P.rvs (size=100000)
rvs
```

```python
plt.hist(rvs)
plt.show()
```

```python
N = stats.norm()
rvs = N.rvs(size = 100000)
plt.hist(rvs, bins=1000)
plt.show()
```
